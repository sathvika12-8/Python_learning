```python
#Student Report card
class StudentReportCard:
  def __init__(self):
    self.student_names=[]
    self.subjects=set()
    self.student_info=()
    self.subject_marks={}
  def collect_student_data(self):
    name=input("Enter student name: ")
    age=int(input("Enter student age: "))
    num_subjects=int(input("Enter number of subjects: "))
    print("\n---Enter subject details---")
    for i in range(num_subjects):
      sub_name=input(f"Enter subject {i+1} name: ")
      marks=float(input(f"Enter marks for {sub_name}: "))
      self.subjects.add(sub_name)
      self.subject_marks[sub_name]=marks
      self.student_names.append(name)
      self.student_info=(name,age)
  def display_report_card(self):
    total_subjects=len(self.subject_marks)
    if total_subjects==0:
      print("No student record found")
      return
    total_marks=sum(self.subject_marks.values())
    average_marks=total_marks/total_subjects
    print("\n---Student Report Card---")
    print(f"Name: {self.student_info[0]}")
    print(f"Age: {self.student_info[1]}")
    print(f"subject:{' ,'.join(self.subjects)}")
    print("----------------------")
    print("Marks Breakdown")
    for subject,marks in self.subject_marks.items():
      print(f"{subject}: {marks}")
    print("----------------------")
    print(f"Total Marks: {total_marks}")
    print(f"Average Marks: {average_marks:.2f}")
    print("----------------------")
if __name__=="__main__":
  report_card=StudentReportCard()
  report_card.collect_student_data()
  report_card.display_report_card()
```
