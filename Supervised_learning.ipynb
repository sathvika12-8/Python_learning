```python
#1.Linear Regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([30,40,50,60,70])

model = LinearRegression()
model.fit(X, y)

plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression")
plt.show()
```

```python
#2.Logistic Regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([20,25,30,35,40]).reshape(-1,1)
y = np.array([0,0,1,1,1])

model = LogisticRegression()
model.fit(X, y)

X_test = np.linspace(20,40,100).reshape(-1,1)
y_prob = model.predict_proba(X_test)[:,1]

plt.plot(X_test, y_prob)
plt.xlabel("Age")
plt.ylabel("Purchase Probability")
plt.title("Logistic Regression")
plt.show()
```

```python
#3.Decision Tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
import numpy as np

X = [[25,50000],[35,65000],[45,80000]]
y = [0,1,1]

model = DecisionTreeClassifier()
model.fit(X,y)

plt.figure(figsize=(8,6))
plot_tree(model, filled=True)
plt.title("Decision Tree")
plt.show()
```

```python
#4.Random Forest
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import numpy as np

X = [[25,50000],[35,65000],[45,80000],[30,60000]]
y = [0,1,1,0]

model = RandomForestClassifier()
model.fit(X,y)

features = ["Age","Salary"]
importance = model.feature_importances_

plt.bar(features, importance)
plt.title("Random Forest Feature Importance")
plt.show()
```

```python
#5.KNN
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X = np.array([[1,2],[2,3],[3,4],[6,7]])
y = [0,0,1,1]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

plt.scatter(X[:,0], X[:,1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("KNN Classification")
plt.show()
```

```python
#6.SVM
import matplotlib.pyplot as plt
from sklearn.svm import SVC
import numpy as np

X = np.array([[1,2],[2,3],[4,5],[6,7]])
y = [0,0,1,1]

model = SVC(kernel='linear')
model.fit(X,y)

plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM Classification")
plt.show()
```

```python
#7.Naive's baye
import matplotlib.pyplot as plt

labels = ['Buy', 'Not Buy']
sizes = [65, 35]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Naive Bayes Class Probability")
plt.show()
```

```python
#8.Polynomial regression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([10,20,40,80,160])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly,y)

plt.scatter(X,y)
plt.plot(X,model.predict(X_poly))
plt.title("Polynomial Regression")
plt.show()
```

```python
#write a program on detecting whether mail is spam or not
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Sample dataset
emails = [
    "Congratulations! You won a lottery",
    "Claim your free prize now",
    "Meeting at 10 AM tomorrow",
    "Project submission deadline extended",
    "Win cash rewards instantly",
    "Let's have lunch today"
]

# Labels (1 = Spam, 0 = Not Spam)
labels = [1, 1, 0, 0, 1, 0]

# Convert text into numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test email
new_email = ["Congratulations! Claim your free cash prize"]

# Transform and predict
new_email_vector = vectorizer.transform(new_email)
prediction = model.predict(new_email_vector)

# Output
if prediction[0] == 1:
    print("This mail is SPAM")
else:
    print("This mail is NOT SPAM")
```
