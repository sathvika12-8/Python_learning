```python
#pandas
#data frame
import pandas as pd

# Creating DataFrame
data = {
    "ID": [1, 2, 3, 4],
    "Name": ["Ram", "Sam", "Tom", "John"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88],
    "Department": ["CSE", "ECE", "CSE", "EEE"]
}

df = pd.DataFrame(data)

# Display DataFrame
print("Original DataFrame:")
print(df)

# Shape
print("\nShape:")
print(df.shape)

# Columns
print("\nColumns:")
print(df.columns)

# Index
print("\nIndex:")
print(df.index)

# Head
print("\nHead:")
print(df.head(2))

# Tail
print("\nTail:")
print(df.tail(2))

# Info
print("\nInfo:")
df.info()

# Describe
print("\nDescribe:")
print(df.describe())

# Select column
print("\nSelect Name column:")
print(df["Name"])

# Select row using loc
print("\nFirst Row using loc:")
print(df.loc[0])

# Select row using iloc
print("\nSecond Row using iloc:")
print(df.iloc[1])

# Sorting
print("\nSort by Marks:")
print(df.sort_values("Marks"))

# Filtering
print("\nMarks greater than 80:")
print(df[df["Marks"] > 80])

# Add new column
df["Grade"] = ["A", "A+", "B", "A"]
print("\nAfter adding Grade:")
print(df)

# Update values
df["Marks"] = df["Marks"] + 5
print("\nAfter updating Marks:")
print(df)

# Drop column
df.drop("Grade", axis=1, inplace=True)
print("\nAfter dropping Grade:")
print(df)

# Check null values
print("\nNull values:")
print(df.isnull())

# Fill null values
df["Marks"].fillna(0, inplace=True)

# Rename column
df.rename(columns={"Marks": "Score"}, inplace=True)
print("\nAfter renaming Marks to Score:")
print(df)

# Sum
print("\nTotal Score:")
print(df["Score"].sum())

# Mean
print("\nAverage Score:")
print(df["Score"].mean())

# Max
print("\nMaximum Score:")
print(df["Score"].max())

# Min
print("\nMinimum Score:")
print(df["Score"].min())

# Count
print("\nCount:")
print(df["Score"].count())

# Unique values
print("\nUnique Departments:")
print(df["Department"].unique())

# Value counts
print("\nDepartment counts:")
print(df["Department"].value_counts())

# GroupBy
print("\nGroup by Department:")
print(df.groupby("Department")["Score"].sum())

# Concatenate
df2 = pd.DataFrame({
    "ID": [5],
    "Name": ["David"],
    "Age": [23],
    "Score": [95],
    "Department": ["CSE"]
})

combined = pd.concat([df, df2])
print("\nConcatenated DataFrame:")
print(combined)

# Merge
extra = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "City": ["Hyderabad", "Delhi", "Mumbai", "Chennai"]
})

merged = pd.merge(df, extra, on="ID")
print("\nMerged DataFrame:")
print(merged)

# Save to CSV
df.to_csv("students.csv", index=False)
print("\nData saved to students.csv")

# Read CSV
new_df = pd.read_csv("students.csv")
print("\nRead CSV file:")
print(new_df)
```

```python
#series
import pandas as pd

# Creating Series
s = pd.Series([10, 20, 30, 40, 20, 10, None])

# Display Series
print("Original Series:")
print(s)

# Head
print("\nHead:")
print(s.head())

# Tail
print("\nTail:")
print(s.tail())

# Shape
print("\nShape:")
print(s.shape)

# Index
print("\nIndex:")
print(s.index)

# Data type
print("\nData Type:")
print(s.dtype)

# Sum
print("\nSum:")
print(s.sum())

# Mean
print("\nMean:")
print(s.mean())

# Max
print("\nMaximum:")
print(s.max())

# Min
print("\nMinimum:")
print(s.min())

# Count (ignores null)
print("\nCount:")
print(s.count())

# Unique values
print("\nUnique values:")
print(s.unique())

# Value counts
print("\nValue counts:")
print(s.value_counts())

# Sort values
print("\nSorted values:")
print(s.sort_values())

# Check null values
print("\nIs Null:")
print(s.isnull())

# Fill null values
print("\nFill Null with 0:")
print(s.fillna(0))

# Drop null values
print("\nDrop Null values:")
print(s.dropna())

# Access element
print("\nFirst element:")
print(s[0])

# Access multiple elements
print("\nElements from index 1 to 3:")
print(s[1:4])

# Apply function
print("\nMultiply each element by 2:")
print(s.apply(lambda x: x * 2 if pd.notnull(x) else x))

# Check duplicates
print("\nDuplicate values:")
print(s.duplicated())

# Remove duplicates
print("\nAfter removing duplicates:")
print(s.drop_duplicates())

# Convert to list
print("\nConvert to List:")
print(s.tolist())

# Convert to DataFrame
print("\nConvert Series to DataFrame:")
print(s.to_frame(name="Numbers"))

# Describe statistics
print("\nDescribe:")
print(s.describe())

# Save to CSV
s.to_csv("series.csv", index=False)
print("\nSeries saved to series.csv")

# Read CSV
new_s = pd.read_csv("series.csv")
print("\nRead CSV:")
print(new_s)
```
