```python
#Matplot lib
#line chart
import matplotlib.pyplot as plt
marks=[98,93,62,78,35,22]
amount=[5000,4500,2500,3000,1000,0]
plt.plot(marks,amount)
plt.xlabel('marks')
plt.ylabel('amount')
plt.title('marks vs amount')
plt.show()
```

```python
#Bar chart
students = ["Ram", "Sam", "Tom", "John"]
marks = [85, 90, 78, 88]
plt.bar(students, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks Bar Chart")
plt.show()
```

```python
#histogram
ages=[18,19,20,21,22,22,23,24,25,25,25]
plt.hist(ages,bins=5)
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('histogram')
plt.show()
```

```python
#scatterplot
hours=[1,2,3,4,5,6]
marks=[30,40,50,60,75,90]
plt.scatter(hours,marks)
plt.title("study hours vs marks")
plt.xlabel(" hours")
plt.ylabel("marks")
plt.show()
```

```python
#piechart
labels=['python','Java','c++','Javascript']
values=[40,25,20,15]
plt.pie(values,labels=labels,autopct='%1.1f%%')
plt.title('programming languages')
plt.show()
```
