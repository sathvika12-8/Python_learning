# **5.Date Feature Extraction**

```python
import pandas as pd

# Create a synthetic PurchaseDate column
start_date = '2023-01-01'
df['PurchaseDate'] = pd.to_datetime(start_date) + pd.to_timedelta(range(len(df)), unit='D')

print("DataFrame with new PurchaseDate column:")
display(df.head())
```

Now that we have a date column, let's extract some useful features from it:

```python
# Extract Year, Month, Day, DayOfWeek, and DayOfYear
df['PurchaseYear'] = df['PurchaseDate'].dt.year
df['PurchaseMonth'] = df['PurchaseDate'].dt.month
df['PurchaseDay'] = df['PurchaseDate'].dt.day
df['PurchaseDayOfWeek'] = df['PurchaseDate'].dt.dayofweek  # Monday=0, Sunday=6
df['PurchaseDayOfYear'] = df['PurchaseDate'].dt.dayofyear

# Extract WeekOfYear and Quarter
df['PurchaseWeekOfYear'] = df['PurchaseDate'].dt.isocalendar().week.astype(int)
df['PurchaseQuarter'] = df['PurchaseDate'].dt.quarter

# Check if it's a weekend
df['IsWeekend'] = (df['PurchaseDate'].dt.dayofweek >= 5).astype(int)

print("DataFrame after extracting date features:")
display(df.head())
```

# **6.Binning**

```python
#convert Age into Categories
bins = [0, 25, 45, df['Age'].max() + 1] # Ensure the last bin covers the maximum age
labels=["Young","Adult","Senior"]
df["Age_Group"]=pd.cut(df["Age"], bins=bins, labels=labels)
df
```

# **7.Outlier Detection**

### Outlier Detection using IQR Method for 'AnnualIncome'

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'AnnualIncome'
Q1 = df['AnnualIncome'].quantile(0.25)
Q3 = df['AnnualIncome'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers_iqr = df[(df['AnnualIncome'] < lower_bound) | (df['AnnualIncome'] > upper_bound)]

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")

print("\nOutliers in 'AnnualIncome' using IQR method:")
display(outliers_iqr)
```

# **8.Feature Interaction**

```python
import numpy as np

# Create a synthetic 'Experience' column for demonstration purposes
# Assuming experience can range from 0 to 40 years
df['Experience'] = np.random.randint(0, 41, size=len(df))

print("DataFrame with synthetic 'Experience' column:")
display(df.head())
```

```python
# Create the new feature: experience_salary
df['experience_salary'] = df['Experience'] * df['AnnualIncome']

print("DataFrame with 'experience_salary' feature:")
display(df.head())
```

### Importance of Feature Interaction: `experience_salary`

Feature interaction involves combining two or more features to create a new, more expressive feature. The `experience_salary` feature, created by multiplying `Experience` and `AnnualIncome`, can be particularly insightful for several reasons:

1.  **Capturing Synergistic Effects:** It captures the combined effect of an individual's experience and their annual income. For instance, a highly experienced person with a high annual income might have different purchasing behaviors or customer values than a highly experienced person with a low income, or a less experienced person with a high income.

2.  **Enhanced Predictive Power:** In many machine learning models, features are treated independently. By explicitly creating interaction terms, we allow the model to understand complex, non-linear relationships that might exist between features. This can significantly improve the model's predictive accuracy for tasks like customer segmentation, churn prediction, or estimating lifetime value.

3.  **Domain-Specific Insights:** In a business context, `experience_salary` could represent an individual's career progression, financial stability, or purchasing power influenced by their professional journey. It can help identify segments of customers who are both experienced and financially well-off, which might be a key demographic for certain products or services.

4.  **Addressing Model Limitations:** Simple linear models might not inherently detect that the effect of `Experience` on a target variable changes depending on the `AnnualIncome`. An interaction term explicitly provides this information to the model.

By creating `experience_salary`, we're essentially providing a more nuanced perspective on the customer's profile, which can lead to better model performance and deeper business understanding.
