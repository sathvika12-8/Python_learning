```python
import pandas as pd

#Load the uploaded file
df = pd.read_csv(file_name)
df
```

```python
#data cleaning operations on rainfall_data.csv
df.head()
df.info()
df.isnull().sum()
```

```python
#convert data column
df["Month"] = df["Month"].astype(str).str.strip()
df["Month"] = pd.to_datetime(df["Month"],errors="coerce")
df.info()
```

```python
#Handling missing values

#Numerical columns
num_cols = ["Relative Humidity","Temperature","Precipitation"]
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

#Fill numerical missing values with mode
df["Specific Humidity"] = df["Specific Humidity"].fillna(df["Specific Humidity"].mode()[0])
#check missing values
print(df.isnull().sum())
```

```python
#Fix text inconsistency
df["Specific Humidity"] = df["Specific Humidity"].astype(str).str.strip()
df["Specific Humidity"].unique()
df
```

### Fixing the 'Month' column and creating a proper 'Date' column

As previously discussed, the `Month` column was converted to `NaT` because it only contained month numbers and not a full date string. To rectify this, we will combine the 'Year', 'Month', and 'Day' columns to construct a complete 'Date' column and then convert it to a datetime object. Subsequently, we'll remove the original 'Year', 'Month', and 'Day' columns to streamline the DataFrame.

```python
# Reload the DataFrame to get original 'Year', 'Month', 'Day' columns
# as they might have been altered or corrupted by previous operations.
# This ensures we start with clean integer values for date components.
df = pd.read_csv(file_name)

# Ensure 'Month' and 'Day' are integers
# (though after reloading, they should already be int, this is a good safeguard)
df['Month'] = df['Month'].astype(int)
df['Day'] = df['Day'].astype(int)

# Create a new 'Date' column by combining Year, Month, and Day
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])

# Drop the original Year, Month, and Day columns
df = df.drop(columns=['Year', 'Month', 'Day'])

# Display the updated DataFrame info and head to verify
df
```

```python
df.to_csv("cleaned_rainfall.csv")
```
