```python
import pandas as pd
import numpy as np

data ={
    "name" : ["Amit","Sara","John","Amit",None],
    "age" : [20,21,np.nan,20,19],
    "gender" : ["Male","Female","Male","Male","Female"],
    "marks":[85,90,105,85,-5],
    "attendance":[90,85,80,90,None],
    "city":["Mumbai","Delhi","mumbai","Mumbai","Pune"]
}
df = pd.DataFrame(data)
df
```

```python
df.info()
df.isnull().sum()
df.describe()
```

```python
#Handling missing values
df["age"].fillna(df["age"].mean(), inplace=True)
df["attendance"].fillna(df["attendance"].mean(), inplace=True)
df["name"].fillna("Unknown",inplace=True)
df
```

```python
df["gender"]=df["gender"].astype(str).str.lower()
df["city"]=df["city"].astype(str).str.lower()
df
```

```python
#fix invalid data
#marks should be --between 0 and 100
df =df[(df["marks"]>=0) & (df["marks"]<=100)]
df
```

```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Check uploaded file name
file_name = list(uploaded.keys())[0]
print("Uploaded file:", file_name)

# Load CSV file
df = pd.read_csv(file_name)

# -----------------------------
# 1. Display first 5 rows
# -----------------------------
print("First 5 rows:")
print(df.head())

# -----------------------------
# 2. Display last 5 rows
# -----------------------------
print("\nLast 5 rows:")
print(df.tail())

# -----------------------------
# 3. Total rows and columns
# -----------------------------
print("\nShape of dataset:")
print(df.shape)

rows, cols = df.shape
print("Rows:", rows)
print("Columns:", cols)

# -----------------------------
# 4. Data types
# -----------------------------
print("\nData types:")
print(df.dtypes)

# -----------------------------
# 5. Missing values
# -----------------------------
print("\nMissing values count:")
print(df.isnull().sum())

# Columns with missing values
print("\nColumns having missing values:")
print(df.columns[df.isnull().any()])

# -----------------------------
# 6. Fill missing values
# Numerical → Mean
# -----------------------------
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mean())

# -----------------------------
# 7. Fill missing values
# Categorical → Mode
# -----------------------------
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

# -----------------------------
# 8. Remove duplicates
# -----------------------------
df = df.drop_duplicates()

print("\nDuplicates removed")

# -----------------------------
# 9. Convert date column
# Example create date column
# -----------------------------
df['date'] = pd.to_datetime("2024-01-01")

print("\nDate column:")
print(df['date'].head())

# -----------------------------
# 10. Rename columns lowercase
# -----------------------------
df.columns = df.columns.str.lower()

print("\nLowercase columns:")
print(df.columns)

# -----------------------------
# 11. Convert float to int
# Example fare column
# -----------------------------
df['fare'] = df['fare'].astype(int)

print("\nUpdated data types:")
print(df.dtypes)

# -----------------------------
# Final cleaned data
# -----------------------------
print("\nCleaned dataset:")
print(df.head())
```

```python
import pandas as pd

from google.colab import files

uploaded = files.upload()

# Check uploaded file name
file_name = list(uploaded.keys())[0]
print("Uploaded file:", file_name)

# Load dataset
df = pd.read_csv(file_name)

# Display first 5 rows
print("First 5 rows:")
print(df.head())

# --------------------------------
# 1. Select multiple columns
# --------------------------------
print("\nSelected columns:")
print(df[['age', 'studytime', 'G3']])

# --------------------------------
# 2. Fill missing values together
# (numerical columns)
# --------------------------------
df[['age', 'studytime', 'G3']] = df[
    ['age', 'studytime', 'G3']
].fillna(df[['age', 'studytime', 'G3']].mean())

print("\nMissing values filled")

# --------------------------------
# 3. Increase marks by 10%
# --------------------------------
df['G3'] = df['G3'] * 1.10

print("\nMarks after 10% increase:")
print(df['G3'].head())

# --------------------------------
# 4. Fix inconsistent text
# Example using address column
# --------------------------------
df['address'] = df['address'].str.lower()

print("\nFixed text in address column:")
print(df['address'].unique())

# --------------------------------
# 5. Create Pass/Fail column
# Pass if marks >= 10
# --------------------------------
df['Result'] = df['G3'].apply(
    lambda x: 'Pass' if x >= 10 else 'Fail'
)

print("\nPass/Fail:")
print(df[['G3', 'Result']].head())

# --------------------------------
# 6. Filter students with marks > 15
# --------------------------------
top_students = df[df['G3'] > 15]

print("\nStudents with marks > 15:")
print(top_students[['age', 'G3', 'Result']])

# --------------------------------
# 7. Drop unnecessary columns
# Example: guardian
# --------------------------------
df = df.drop(columns=['guardian'])

print("\nColumn 'guardian' removed")

# --------------------------------
# 8. One-Hot Encoding
# Example on school and sex
# --------------------------------
df = pd.get_dummies(df, columns=['school', 'sex'])

print("\nAfter one-hot encoding:")
print(df.head())

# --------------------------------
# 9. Sort dataset by marks
# --------------------------------
sorted_df = df.sort_values(by='G3', ascending=False)

print("\nSorted by marks:")
print(sorted_df[['age', 'G3']].head())
```
