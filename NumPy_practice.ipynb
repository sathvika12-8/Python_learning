```python
#numpy
import numpy as np

# Array creation
a = np.array([1, 2, 3, 4, 5, 6])
print("Original Array:", a)

# zeros()
zeros_array = np.zeros((2, 3))
print("\nZeros Array:\n", zeros_array)

# ones()
ones_array = np.ones((2, 2))
print("\nOnes Array:\n", ones_array)

# arange()
arange_array = np.arange(1, 10, 2)
print("\nArange:", arange_array)

# linspace()
linspace_array = np.linspace(1, 10, 5)
print("\nLinspace:", linspace_array)

# shape
print("\nShape of a:", a.shape)

# reshape()
reshaped = a.reshape(2, 3)
print("\nReshaped Array:\n", reshaped)

# flatten()
flattened = reshaped.flatten()
print("\nFlattened Array:", flattened)

# sum()
print("\nSum:", np.sum(a))

# mean()
print("Mean:", np.mean(a))

# max()
print("Maximum:", np.max(a))

# min()
print("Minimum:", np.min(a))

# sqrt()
print("Square Root:", np.sqrt(a))

# std()
print("Standard Deviation:", np.std(a))

# var()
print("Variance:", np.var(a))

# sort()
unsorted = np.array([4, 2, 1, 6, 3, 5])
print("\nSorted Array:", np.sort(unsorted))

# concatenate()
b = np.array([7, 8, 9])
concatenated = np.concatenate((a, b))
print("\nConcatenated Array:", concatenated)

# split()
split_array = np.split(a, 3)
print("\nSplit Array:", split_array)

# random.rand()
print("\nRandom Float Values:", np.random.rand(3))

# random.randint()
print("Random Integer Values:", np.random.randint(1, 10, 5))
```
