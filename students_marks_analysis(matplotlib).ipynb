```python
#challenge-student marks analysis
import matplotlib.pyplot as plt
students=['A','B','C','D','E','F','G','H','I','J']
marks=[78,85,92,67,88,95,72,81,90,76]
plt.bar(students,marks)
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Student Marks Analysis')
plt.show()
highest=max(marks)
index=marks.index(highest)
print("Highest scorer",students[index])
print("marks",highest)
sorted_marks=sorted(marks,reverse=True)
print(sorted_marks)
plt.bar(range(len(sorted_marks)),sorted_marks)
plt.title("marks in descending order")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```
